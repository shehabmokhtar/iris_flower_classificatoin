```python
# Importing libararies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read the dataset
dataset = pd.read_csv('IRIS.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
# Top 5 rows of the dataset
dataset.head(5)
```

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

#Support Vector Machine SVM Model

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svm_classifier = SVC()
svm_classifier.fit(X_train, y_train)
y_pred_svm = svm_classifier.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print("SVM Accuracy: ", accuracy_svm*100 ,"%")
```

# K-Nearest Neighbors KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier

knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train, y_train)
y_pred_knn = knn_classifier.predict(X_test)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print("\nKNN Accuracy:", accuracy_knn*100,"%")
```

# Naive Bayes Model

```python
from sklearn.naive_bayes import GaussianNB

nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)
y_pred_nb = nb_classifier.predict(X_test)
accuracy_nb = accuracy_score(y_test, y_pred_nb)
print("\nNaive Bayes Accuracy:", accuracy_nb*100,"%")
```

# Predicate a new data

```python
new_data = [[5.1, 3.5, 1.4, 0.2]]

# Predicate a new data ( SVM Model )
predicted_svm = svm_classifier.predict(new_data)
print('SVM Model Predicated data        : ',predicted_svm[0])

# Predicate a new data ( KNN Model )
predicted_knn = knn_classifier.predict(new_data)
print('KNN Model Predicated data        : ',predicted_knn[0])

# Predicate a new data ( Naive Bayes Model )
predicted_nb = nb_classifier.predict(new_data)
print('Naive Bayes Model Predicated data: ',predicted_nb[0])
```

# Create Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Create a confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_svm)

# Plot the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=X_test, yticklabels=y_test)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
